```python
import xarray as xr
import matplotlib.pyplot as plt

npth='J:/MDLS_OBS_OHC/'
```

```python
DATA_REAL=xr.open_dataset(npth+'myIAP_198001_201812_tshuv_real.nc').loc[dict(lat=slice(-80,-30))].OHC2000
DATA=xr.open_dataset(npth+'myIAP_198001_201812_tshuv.nc').loc[dict(lat=slice(-80,-30))].OHC2000

DATA_REAL1=xr.open_dataset(npth+'myIAP_198001_201812_tshuv_real.nc').loc[dict(lat=slice(-60,-50),lon=slice(153,290))].OHC700
DATA1=xr.open_dataset(npth+'myIAP_198001_201812_tshuv.nc').loc[dict(lat=slice(-60,-50),lon=slice(153,290))].OHC700
```

```python
DATA=DATA.where(DATA<10**20)
DATA1=DATA1.where(DATA1<10**20)
```

```python
DATA_REAL.mean(dim=['lat','lon']).plot()
DATA.mean(dim=['lat','lon']).plot()
```

```python
DATA_REAL1.mean(dim=['lat','lon']).plot()
DATA1.mean(dim=['lat','lon']).plot()
```

```python
M=(DATA1.mean(dim=['lat','lon']).mean()+DATA_REAL1.mean(dim=['lat','lon']).mean())/2
diff=DATA1.mean(dim=['lat','lon'])-DATA_REAL1.mean(dim=['lat','lon'])

plt.title('SCP difference (OHC700)')
plt.plot(DATA1.time,DATA_REAL1.mean(dim=['lat','lon']),label='Downloaded')
plt.plot(DATA1.time,DATA1.mean(dim=['lat','lon']),label='Calculated')
plt.plot(DATA1.time, diff +DATA_REAL1.mean(dim=['lat','lon']).mean(),label='diff+Average')
plt.legend()
```
